```python
def StokesLet(r):
    R = sqrt(dot(r,r))
    if R < 1e-6:
        return zeros(len(r),len(r))
    
    return (eye(len(r))/R+r*r.reshape(-1,1)/R**3)/(8*pi)

def StressLet(r):
    R = sqrt(dot(r,r))
    if R < 1e-6:
        return zeros(len(r), len(r), len(r))
    return -r*r.reshape(-1,1,1)*r.reshape(1,-1,1)/(6*pi*R**5)
```

```python
N = 50

x = linspace(-1,1, N)
y = linspace(-1,1, N)

r = zeros((N,N,2))
R = zeros((N,N))
S = zeros((N,N,2,2))
T = zeros((N,N,2,2,2))
for i in xrange(N):
    for j in xrange(N):
        r[i,j,0] = x[i]
        r[i,j,1] = y[j]
        R[i,j] = sqrt(dot(r[i,j,:], r[i,j,:]))
        S[i,j] = StokesLet(r[i,j,:])
        T[i,j] = StressLet(r[i,j,:])
```

```python
b = array([1., 0.])
c = array([1., 0.])
Sb = S.dot(b)
Sc = T.dot(b).dot(c)

ids = (R>.2)

ax = gca()
ax.quiver(r[ids,0], r[ids,1], Sb[ids,0], Sb[ids,1],angles='xy',scale_units='xy',scale=2)
ax.set_xlim([-1,1])
ax.set_ylim([-1,1])
axis('equal')
```

```python

ax = gca()
ax.quiver(r[:,:,0], r[:,:,1], Sc[:,:,0], Sc[:,:,1],angles='xy',scale_units='xy',scale=2)
ax.set_xlim([-1,1])
ax.set_ylim([-1,1])
axis('equal')
```

```python
ax = gca()
ax.quiver(r[:,:,0], r[:,:,1], (Sb+Sc)[:,:,0], (Sb+Sc)[:,:,1],angles='xy',scale_units='xy',scale=1)
ax.set_xlim([-1,1])
ax.set_ylim([-1,1])
axis('equal')
```
